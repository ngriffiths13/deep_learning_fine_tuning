```python
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tuned_train_losses = joblib.load('../data/tuned_train_losses.pkl')
tuned_val_losses = joblib.load('../data/tuned_val_losses.pkl')
pretrain_train_losses = joblib.load('../data/pretrain_train_losses.pkl')
pretrain_val_losses = joblib.load('../data/pretrain_val_losses.pkl')
no_pretrain_train_losses = joblib.load('../data/no_pretrain_train_losses.pkl')
no_pretrain_val_losses = joblib.load('../data/no_pretrain_val_losses.pkl')
```

```python
f, ax = plt.subplots(figsize=(8,10))
tu_tr = sns.lineplot(range(len(tuned_train_losses)), tuned_train_losses)
tu_val = sns.lineplot(range(len(tuned_val_losses)), tuned_val_losses)
p_tr = sns.lineplot(range(len(pretrain_train_losses)), pretrain_train_losses)
p_val = sns.lineplot(range(len(pretrain_val_losses)), pretrain_val_losses)
r_tr = sns.lineplot(range(len(no_pretrain_train_losses)), no_pretrain_train_losses)
r_val = sns.lineplot(range(len(no_pretrain_val_losses)), no_pretrain_val_losses)
plt.legend(['tuned_train', 'tuned_val', 'transfer_train', 'transfer_val', 'reg_train', 'reg_val'])
tu_val.lines[0].set_linestyle("--")
p_val.lines[0].set_linestyle("--")
r_val.lines[0].set_linestyle("--")
```
